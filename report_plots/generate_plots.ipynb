### AirSim Grid Generation

```python
import random
import numpy as np
import math

x_min, x_max = -8, 2
y_min, y_max = -8, 0
z_min, z_max = -4, -2
res_z = 2
res_y = 2
res_x = 2
res_yaw = 2
x_vals = np.arange(x_min, x_max+ 1, res_x)
y_vals = np.arange(y_min, y_max+ 1, res_y)
z_vals = np.arange(z_min, z_max + 1, res_z)
yaw_vals = np.arange(0, int(2*math.pi), res_yaw)
num_of_points = 0
points_x = []
points_y = []
points_z = []

total_grid = [[[None for _ in z_vals] for _ in y_vals] for _ in x_vals]
for i, x in enumerate(x_vals):
    for j, y in enumerate(y_vals):
        for k, z in enumerate(z_vals):
            random_start_yaw = random.random() * 2*math.pi
            points = []
            for yaw_shift in yaw_vals:
                yaw = random_start_yaw + yaw_shift
                yaw = yaw if yaw < 2*math.pi else yaw - 2*math.pi
                points_x.append(x)
                points_y.append(y)
                points_z.append(z)
                points.append((x, y, z, int(math.degrees(yaw))))
                num_of_points += 1
            total_grid[i][j][k] = points
```

```python
%matplotlib ipympl
plt.figure()
ax = plt.axes(projection= '3d')
ax.scatter3D(points_x, points_y, points_z, label = "Waypoints", color = 'blue')
ax.set_zlim([-1, -5])
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
ax.set_zlabel("z (m)")
ax.set_title("Grid waypoints")
ax.legend()
plt.show()
```

```python
from itertools import product
def get_trajectory_and_update_grid(total_grid, x_vals, y_vals, z_vals):
    is_start_point_found = False
    while not is_start_point_found:
        start_x, start_y, start_z = random.randint(0, len(x_vals)-1), random.randint(0, len(y_vals)-1), random.randint(0, len(z_vals)-1)
        if len(total_grid[start_x][start_y][start_z]) > 0:
            is_start_point_found = True
    path = []
    queue = [(start_x, start_y, start_z)]
    while queue:
        x, y, z = queue.pop(0)
        if len(total_grid[x][y][z]) > 0:
            path.append(total_grid[x][y][z][0])
            del total_grid[x][y][z][0]

            all_candidates = list(product([x - 1, x, x +1],[y - 1, y, y + 1],[z - 1, z, z + 1]))
            valid_candidates = []
            for i in all_candidates:
                x, y, z = i
                if x >= 0 and x < len(x_vals) and \
                   y >= 0 and y < len(y_vals) and \
                   z >= 0 and z < len(z_vals) and \
                   len(total_grid[x][y][z]) > 0:
                    valid_candidates.append((x, y, z))
            
            if valid_candidates:
                queue.append(random.sample(valid_candidates, k= 1)[0])
            
    return path
path = get_trajectory_and_update_grid(total_grid, x_vals, y_vals, z_vals)
```

```python
trajectory = np.array(path)
traj_x = trajectory[:,0]
traj_y = trajectory[:,1] 
traj_z = trajectory[:,2] 
plt.figure()
ax = plt.axes(projection= '3d')
ax.plot3D(traj_x, traj_y, traj_z, label = "Trajectory", color = 'red')
ax.set_zlim([-1, -5])
ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")
ax.set_zlabel("z (m)")
ax.set_title("Training Trajectory")
ax.legend()
plt.show()
```

```python
def get_random_x_y_alt():
    return [random.uniform(x_min, x_max), random.uniform(y_min, y_max), random.uniform(z_min, z_max)]

random_x_y_alt = np.array([get_random_x_y_alt() for i in range(15)])
```
