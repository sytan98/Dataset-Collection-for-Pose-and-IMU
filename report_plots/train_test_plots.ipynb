```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="whitegrid")
```

## Baseline

### Noisy None

```python
noisy_none_train = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_None-tag-Loss_train_total.csv')
noisy_none_test = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_None-tag-Loss_val_total.csv')
noisy_none_test.drop(5,inplace=True)
```

```python
plt.plot(noisy_none_train["Step"], noisy_none_train["Value"], label="Train")
plt.plot(noisy_none_test["Step"], noisy_none_test["Value"], label="Test")
plt.xlabel('Epoch')
plt.ylabel('Absolute Error')
plt.legend()
```

```python
noisy_v1_train = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_v1-tag-Loss_train_total.csv')
noisy_v1_test = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_v1-tag-Loss_val_total.csv')
noisy_v1_test.drop(5,inplace=True)
```

```python
noisy_v2_train = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_v2-tag-Loss_train_total.csv')
noisy_v2_test = pd.read_csv('./data/run-runs_AirSim_building_new_mapnet_mapnet_imu_None_noisy_v2-tag-Loss_val_total.csv')
noisy_v2_test.drop(5,inplace=True)
```

```python
plt.plot(noisy_none_train["Step"], noisy_none_train["Value"], label="Original Train")
plt.plot(noisy_none_test["Step"], noisy_none_test["Value"], label="Original Test")
plt.plot(noisy_v1_train["Step"], noisy_v1_train["Value"], label="Noisy_v1 Train")
plt.plot(noisy_v1_test["Step"], noisy_v1_test["Value"], label="Noisy_v1 Test")
plt.plot(noisy_v2_train["Step"], noisy_v2_train["Value"], label="Noisy_v2 Train")
plt.plot(noisy_v2_test["Step"], noisy_v2_test["Value"], label="Noisy_v2 Test")
plt.xlabel('Epoch')
plt.ylabel('Absolute Error')
plt.ylim(-10,20)
plt.xlim(0,100)
plt.legend()
```

```python
plt.plot(noisy_none_test["Step"], noisy_none_test["Value"], label="Original Test")
plt.plot(noisy_v1_test["Step"], noisy_v1_test["Value"], label="Noisy_v1 Test")
plt.plot(noisy_v2_test["Step"], noisy_v2_test["Value"], label="Noisy_v2 Test")
plt.xlabel('Epoch')
plt.ylabel('Absolute Error')
plt.xlim(75, 100)
plt.ylim(1.4, 2.0)
plt.legend()
```
