```python
# Importing all necessary libraries 
import cv2 
import os 

# Read the video from specified path 
cam = cv2.VideoCapture("C:/Users/tansi/Documents/Imperial_College_London/Year3/FYP_Project/School_trajectory.MOV") 
image_dir = 'D:/Imperial/FYP/captured_data/campus_10fps'
try: 
	
	# creating a folder named data 
	if not os.path.exists(image_dir): 
		os.makedirs(image_dir) 

# if not created then raise error 
except OSError: 
	print ('Error: Creating directory of data') 

# frame 
currentframe = 0

while(True): 

    # reading from frame 
    ret,frame = cam.read() 

    if ret: 
        # if video is still left continue creating images 
        name = image_dir +'/frame' + str(currentframe) + '.jpg'

        # writing the extracted images 
        if currentframe % 3 == 0:
            print ('Creating...' + name) 
            cv2.imwrite(name, frame) 

        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```
